```python
# WORKS FROM PYTHON !!!!!

robjects.r('''someData = dk_3d %>%
  filter(surf == "inflated" & hemi == "right") %>%
  unnest(ggseg_3d) %>%
  ungroup() %>%
  select(region) %>%
  na.omit() %>%
  mutate(p = sample(seq(0,.5, length.out = 100 ), nrow(.)) %>%
           round(2))

p = ggseg3d(.data = someData,
        atlas = dk_3d,
        colour = "p", text = "p") %>%
  pan_camera("right medial")
  
saveWidget(p, "html_outputs/robjects_example.html", selfcontained = F)
  
print(p)
''')
```

```python
# LEFT HEMISPHERE

robjects.r('''colors = c("indianred4",'indianred2','coral1','lightpink1','mistyrose1')
p = ggseg3d(.data = dk_data, 
        atlas = dk_3d,
        hemisphere = 'left',
        colour = "p", 
        palette = colors,
        text = "p") %>% 
  pan_camera("right medial")
  
  print(p)
''')
```

```python
# RIGHT HEMISPHERE

robjects.r('''colors = c("indianred4",'indianred2','coral1','lightpink1','mistyrose1')
p = ggseg3d(.data = dk_data, 
        atlas = dk_3d,
        hemisphere = 'right',
        colour = "p", 
        palette = colors,
        text = "p") %>% 
  pan_camera("left medial")
  
  print(p)
''')
```

```python
# WHOLE BRAIN

robjects.r('''colors = c("indianred4",'indianred2','coral1','lightpink1','mistyrose1')
p = ggseg3d(.data = dk_data, 
        atlas = dk_3d,
        hemisphere = c('left','right'),
        colour = "p", 
        palette = colors,
        text = "p",
        options.legend = list(title=list(text=""))) %>% 
  pan_camera("left medial")
  
  print(p)
''')
```

## ASEG-atlas

```python
robjects.r('''
aseg_data <- read.csv(file = 'aseg_R_subtype0.csv')
''')
```

```python
# WITHOUT REFERENCE

robjects.r('''colors = c("indianred4",'indianred2','coral1','lightpink1','mistyrose1')
p = ggseg3d(.data = aseg_data, 
        atlas = aseg_3d,
        colour = "p", 
        palette = colors,
        text = "p",
        options.legend = list(title=list(text="")))
  
  print(p)
''')
```

```python
# ADD GLASSBRAIN
# issue: only glassbrain() rotates in plotly

robjects.r('''colors = c("indianred4",'indianred2','coral1','lightpink1','mistyrose1')
p = ggseg3d(.data = aseg_data, 
        atlas = aseg_3d,
        colour = "p", 
        palette = colors,
        text = "p",
        na.alpha= .5,
        options.legend = list(title=list(text=""))) %>% 
  add_glassbrain()
  
  print(p)
''')
```
